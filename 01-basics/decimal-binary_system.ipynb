Aim

The aim of this program is to understand how to convert a decimal number into a binary number and how to convert a binary number back into a decimal number using C++.

Theory

A computer understands only 0 and 1. Therefore, every decimal number is internally stored as a binary number.

For example,

Decimal 5 = Binary 101

Decimal 10 = Binary 1010

There are two common methods to convert decimal into binary.

Division Method
Bitwise Method

```python
Method 1 : Decimal to Binary (Division Method)
Algorithm

Step 1:
Take a decimal number.

Step 2:
Divide it by 2.

Step 3:
Store the remainder.

Step 4:
Repeat until the quotient becomes 0.

Step 5:
Reverse the stored remainders.

The reversed number is the binary equivalent.
```

Example

Convert 10 into binary.

Division	Remainder
10/2 = 5	0
5/2 = 2	1
2/2 = 1	0
1/2 = 0	1

Remainders = 0101

Reverse = 1010

Answer = 1010

```python
#include<iostream>
using namespace std;

int main(){
    int n;
    cin >> n;

    int ans = 0;
    int i = 0;

    while(n != 0){
        int bit = n % 2;

        ans = ( bit * pow(10, i) ) + ans;

        n = n/2;
        i++;
        
    }

    cout << "Binary is" << ans << endl;

    return 0;
    
}



```

Explanation

The remainder obtained after every division is either 0 or 1.

These bits are generated in reverse order.

Therefore, every new bit is placed according to its position using

bit × 10^i

Finally the complete binary number is obtained.

Method 2 : Decimal to Binary (Bitwise Method)
Theory

Instead of dividing by 2, we can use bitwise operators.

Operator 1
&

It checks the last bit.

bit = n & 1

If the last bit is 1

Output = 1

If the last bit is 0

Output = 0

```python
Operator 2
>>

Right Shift Operator

n >> 1

removes the last bit from the binary representation.

Example

10110

↓

1011
```

```python
#include <iostream>
#include <cmath>

using namespace std;

int main(){

    int n;
    cin >> n;

    int ans = 0;
    int i = 0;

    while(n != 0){
        int bit = n & 1;
        ans = (bit * pow(10,i)) + ans;

        n = n>>1;
        i++;

    }
    cout << "Binary = " << ans;

    return 0;
}
```

Explanation

The operator

n & 1

extracts the last bit.

After storing it,

n >> 1

removes the last bit.

This process continues until the number becomes 0.

Binary to Decimal
Theory

Every binary digit represents a power of 2.

Example

10101

Position

1    0    1    0    1
↑    ↑    ↑    ↑    ↑

2^4 2^3 2^2 2^1 2^0

Now,

1×2⁴ + 0×2³ + 1×2² + 0×2¹ + 1×2⁰

=16+0+4+0+1

=21

```python
#include <iostream>
#include <cmath>

using namespace std;

int main()
{
    int n;
    cin >> n;

    int ans = 0;
    int i = 0;

    while(n != 0)
    {
        int digit = n % 10;

        if(digit == 1)
        {
            ans = ans + pow(2,i);
        }

        n = n / 10;
        i++;
    }

    cout << "Decimal = " << ans;

    return 0;
}
```

Explanation

Each binary digit is multiplied with its corresponding power of 2.

Only digits equal to 1 contribute to the final answer.

Finally, all these values are added to obtain the decimal number.

Output

Input

10

Output

Binary = 1010

Input

10101

Output

Decimal = 21
