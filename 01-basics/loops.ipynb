Write a program that continuously asks the user to enter integers. Keep a running total of the positive numbers. If the user enters a negative number, terminate the loop and print the final sum.

```python
#include<iostream>
using namespaces std

int main(){
    cout << "Enter your number:";
    int n;
    int sum = 0;
    cin >> n >> endl;
    
    while(n>=0){
        sum = sum + n;
        cin >> n >> endl;


    cout << "The total sum is:" << sum << endl;
    return 0;


    }
}
```
