Write a program that continuously asks the user to enter integers. Keep a running total of the positive numbers. If the user enters a negative number, terminate the loop and print the final sum.

```python
#include <iostream>
using namespace std;

int main(){
    cout << "Enter your number:";
    int n;
    int sum = 0;
    cin >> n >> endl;
    
    while(n>=0){
        sum = sum + n;
        cin >> n >> endl;


    cout << "The total sum is:" << sum << endl;
    return 0;


    }
}
```

Ask the user to input a positive integer N. Use a while loop to print all numbers from 1 to N, but append the text "(Even)" or "(Odd)" next to each number using an if-else structure.

```python
#include <iostream>
using namespace std;

int main() { 
    cout << "Enter your number N: ";
    int N;
    cin >> N;  

    if (N > 0) {
        int i = 0; 
        while (i < N) {
            if (i % 2 == 0) {
                cout << i << " even\n"; 
            } else {
                cout << i << " odd\n";
            }
            i = i + 1;
        }
    }
    return 0;
}
```
